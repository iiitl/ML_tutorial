```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
db=pd.read_csv('Salary_Data.csv')
db.info()
```

```python
db.isna().sum()
```

```python
x=db.iloc[:,:-1]
y=db.iloc[:,-1]
print(x)
print("\n\n")
print(y)
```

```python
plt.scatter(x,y, color='orange')
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg= PolynomialFeatures(degree=3)
x_poly=poly_reg.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_poly,y,test_size=0.15,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x_train,y_train)
```

```python
y_pred=lin_reg.predict(x_test)
y_pred
```

```python
y_pred_full=lin_reg.predict(x_poly)
```

```python
plt.scatter(x,y,color='blue',label='Distribution')
plt.plot(x,y_pred_full, label="Polynomial Regression", color="purple")
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
